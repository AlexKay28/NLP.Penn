```python
import pandas as pd

train = pd.read_csv("task3_training.tsv",sep='\t')
train.head()
```

```python
drop_list = [i+9 for i in range(45)]
train.drop(train.columns[drop_list], axis='columns', inplace=True)
train
```

```python
train_ADR = train.dropna()
train_ADR
```

```python
for tweet in enumerate(train_ADR['tweet'].to_list()[:5]):
    print(tweet[0],':', tweet[1], '\n')
```

```python
tweet_l = train_ADR['tweet'].to_list()
extraction_l = train_ADR['extraction'].to_list()
drug_l = train_ADR['drug'].to_list()
```

```python
for com in zip(tweet_l[:5], extraction_l, drug_l ):
    print(str(com[1]).capitalize(),'-', str(com[2]).capitalize())
    print(com[0], '\n')
```
